```python
import pandas as pd 
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("data/cleaned_data.xlsx")
```

```python
# Calculate churn rate by Tech Support
tech_support_churn = (
    df.groupby("Tech Support")["Churn Label"]
      .apply(lambda x: (x == "Yes").mean() * 100)
      .reset_index(name="Churn Rate (%)")
)

tech_support_churn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))

sns.barplot(
    data=tech_support_churn,
    x="Tech Support",
    y="Churn Rate (%)"
)

plt.title("Churn Rate by Tech Support")
plt.xlabel("Tech Support")
plt.ylabel("Churn Rate (%)")

plt.show()
```

I expect customers with Tech Support to have a lower churn rate because they receive assistance when they face technical issues.
